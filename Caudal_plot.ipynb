# GRAFICANDO SERIES DE TIEMPO DE CAUDALES

```python
# Importando las librer√≠as a emplear:
import numpy as np
import pandas as pd
import matplotlib
import os
import hvplot.pandas
import holoviews as hv
```

```python
data=pd.read_excel('./test.xlsx')
data['Date'] = pd.to_datetime(data.Date)
data.set_index('Date', inplace=True)
```

```python
# Separando por tramos, arreglar al gusto
g1=['ID01','ID02','ID03']
g2=['ID04','ID05']
#g3=['ID1844','ID1888','ID1911']
```

```python
# En el margen derecho del gr√°fico hay herramientas para hacer ampliar, mover y guardar
f0=data.hvplot.line(x='Date',y=g1,value_label='Caudal m3/s',title='Serie de Caudal grupo 1',xlabel='FECHA')
f0
```

```python
# Si quieres verlo mes a mes en groupby=[index.month] , en caso quiera a√±o a a√±o cambiar a groupby=[index.year]
# Si quiero m√°s grupos al lado se usa el simbolo + (como este ejemplo)
# Con width se cambia el ancho de la imagen
f1=data.hvplot.line(x='Date',y=g1,value_label='Caudal m3/s',width=400,groupby=['index.month'],title='Caudal ... Grupo 1',xlabel='FECHA')
f2=data.hvplot.line(x='Date',y=g2,value_label='Caudal m3/s',width=400,groupby=['index.month'],title='Caudal ... Grupo 2',xlabel='FECHA')
f1+f2
```

## Otros Tipo de gr√°ficos que tiene esta librer√≠a:
>* .area(): Plots a area chart similar to a line chart except for filling the area under the curve and optionally stacking
>* .bar(): Plots a bar chart that can be stacked or grouped
>* .bivariate(): Plots 2D density of a set of points
>* .box(): Plots a box-whisker chart comparing the distribution of one or more variables
>* .heatmap(): Plots a heatmap to visualizing a variable across two independent dimensions
>* .hexbins(): Plots hex bins
>* .histogram(): Plots the distribution of one or histograms as a set of bins
>* .kde(): Plots the kernel density estimate of one or more variables.
>* .line(): Plots a line chart (such as for a time series)
>* .scatter(): Plots a scatter chart comparing two variables
>* .step(): Plots a step chart akin to a line plot
>* .table(): Generates a SlickGrid DataTable
>* .violin(): Plots a violin plot comparing the distribution of one or more variables using the kernel density estimate

```python
# Cajas por mes
b1=data.hvplot.box(by='Date',y=g1,value_label='Caudal m3/s',groupby=['index.month'],title='Boxplot Grupo 1',xlabel='FECHA')
b1
```
